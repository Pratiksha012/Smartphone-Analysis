```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv('processed_final.csv')
df
```

# Scatter Plot

```python
sns.set(style="white")
sns.scatterplot(df, x="Offer_Price", y="Ratings", hue="Flipkart_Assured", size="Reviews", sizes=(50, 200), marker="o")
plt.legend(bbox_to_anchor = (1,1))
plt.title("Scatter Plot for Offer Price and Ratings")
plt.xlabel("Offer Price")
plt.ylabel("Ratings")
plt.show()
```

#### From the above Scatter Plot, we infer that phones with higher price are lowly rated while products that are low priced are highly rated, we also see there are very less phones that are not assured by flipkart and is represented as blue data points. The size of the phones determine the reviews of the product

```python
sns.scatterplot(data=df, x="Ratings", y="Reviews", hue="5G_Support", s = 100, palette="Set2", marker="D")
plt.legend(title = '5G Support')
plt.title("Scatter Plot for Ratings vs Reviews")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

#### From the plot we infer that products that are highly rated are not 5g supported where as products with less reviews are 5g supported. 

```python
plt.style.use('seaborn-v0_8-white')
sns.scatterplot(data=df, x="Battery_mAh", y="Brand", hue = '5G_Support', s = 100, palette="cool", marker="v", color = ['red','blue'])
plt.title("Scatter Plot for Battery Capacity vs Brand")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Brands")
plt.show()
```

#### From the above scatter plot we infer that only certain phone brands like Xiaomi, Redmi, vivo and other brands support 5g and also most of the phones contain a battery mah of 5000. 

```python
fig = px.scatter(df, x="Ratings", y="Offer_Price", color="Brand",
                 hover_name='Product Name', size = 'Ratings', size_max = 40)
fig.update_layout(title="Interactive plot for Ratings and Offer Price", xaxis_title="Ratings", yaxis_title="Offer Price")
fig.show()
```

#### From the above interactive plot we can see the offer price and the ratings of each phone that is colored based on its brands, we also see that brands with lesser price have a high number of ratings. Redmi has the highest ratings followed by vivo. 

```python
fig = px.scatter(df, x="Stars_out_of_5", hover_name="Product Name",
                 color = 'Stars_out_of_5', size = 'Ratings', size_max = 50)
fig.update_layout(title="Interactive Scatter Plot for products based on stars", xaxis_title="Stars")
fig.show()
```

#### From the above plot we can clearly infer that most of the phones are rated between 4.2 to 4.4 and there are limited phones that have high and low number of stars, the size indicates the number of ratings of each phone, the higher the value, the higher number of ratings. 

```python
fig = px.scatter(df, x = 'Original_Price', color = 'Brand',
                 size = 'Ratings', log_x = True, hover_name = 'Product Name', size_max = 40)
fig.update_layout(title="Interactive Scatter Plot for Original Price", xaxis_title="Original Price")
fig.show()
```

#### This plot summarises the prices of each phone with the colors defining the brand of the phone and the size of the phone depends on the number of ratings received by the phone, we see that the increase in price results in lesser number of ratings inferring that phones that are costly are bought by a smaller number of people. 

# Pie Charts

```python
brand_counts = df['Brand'].value_counts() 
plt.figure(figsize=(15,8)) 
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%',shadow=True,startangle=45) 
plt.title('Highly purchased brand') 
plt.show() 
```

#### This is a pie chart which illustrates the highly purchased brand from the web scraped dataset. From the obtained pie chart, we can infer that the most purchased is â€˜Redmiâ€™ at  15% , second highest being â€˜Samsumgâ€™ at 14.3%. The least of all being â€˜Micromaxâ€™ at 0.3%. It can be interpreted that most of the population is attracted towards affordable phones like Redmi and Samsung which shares the top two positions.

```python
product_sales = df['Product Name'].value_counts() 
top_n = 10  
top_selling_products = product_sales.head(top_n) 
plt.figure(figsize=(10, 6)) 
plt.pie(top_selling_products, labels=top_selling_products.index, autopct='%1.1f%%', startangle=140) 
plt.title('Top Selling Products')  
plt.show()
```

#### This is a pie chart which is plotted to find the top selling phone on flipkart ,  From the following dataset we can infer that people prefer 5G phones currently. The OnePlus Nord CE 3 Lite 5G and REDMI Note 13 5G are among the top-selling products on Flipkart, with each having a percentage of 11.5% and 10.3%, respectively. The OnePlus Nord CE 3 Lite 5G and REDMI Note 13 5G are also the most popular models among the given list of mobile phones, as they have the highest percentage of sales.

```python
discounted_sales = df[df['Offer_Price'] < df['Original_Price']].shape[0] 

non_discounted_sales = df[df['Offer_Price'] >= df['Original_Price']].shape[0] 

  

plt.figure(figsize=(8, 6)) 

plt.pie([discounted_sales, non_discounted_sales], labels=['Discounted Sales', 'Non-Discounted Sales'], autopct='%1.1f%%', startangle=45) 

plt.title('Impact of Discounts on Sales Volume')  

plt.show() 
```

#### The sales of products on discounted price is higher at 95.6% than those on non discount at 4.4%. This shows that discounts has a strong impact on the sales and influences it and customers prefer buying products on discounted price. 

```python
brand_name = 'apple' 
brand_data = df[df['Brand'] == brand_name] 
model_sales = brand_data['Product Name'].value_counts() 
plt.figure(figsize=(10, 6)) 
plt.style.use('seaborn-v0_8-white')
plt.pie(model_sales, labels=model_sales.index, autopct='%1.1f%%', startangle=140,shadow=True) 
plt.title('Distribution of ' + brand_name + ' Models') 
plt.axis('equal')  
plt.show() 
```

# Bar Graphs & Histograms

```python
sorted_df = df.sort_values(by='Stars_out_of_5', ascending=False)
top= sorted_df.head(10)
plt.style.use('seaborn-v0_8')
original_price = top['Original_Price']
offer_price = top['Offer_Price']
indices = np.arange(len(original_price))
bar_width = 0.35
plt.figure(figsize=(12, 6))
plt.bar(indices - bar_width/2, original_price, bar_width, label='Original Price')
plt.bar(indices + bar_width/2, offer_price, bar_width, label='Offer Price')
plt.xlabel('Product')
plt.ylabel('Price')
plt.title('Original Price vs. Offer Price')
plt.xticks(indices, top['Product Name'], rotation=50)
plt.legend()
plt.tight_layout()
plt.show()
```

#### The provided code creates a side-by-side bar plot comparing the original and offer prices for a list of products. Each bar represents a product, with two bars side by side showing the original price and offer price respectively.This plot allows for a visual comparison between the original and offer prices of different products, helping to understand the pricing differences across the product list.

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(10, 6))
average_ratings_by_brand = df.groupby('Brand')['Stars_out_of_5'].mean().sort_values(ascending=False)
plt.bar(average_ratings_by_brand.index, average_ratings_by_brand, color='skyblue')
plt.title('Average Ratings by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=70)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Stars_out_of_5', bins=20, kde=True, color='skyblue')
plt.title('Histogram of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

#### The first plot helps to visualize the average ratings of phones by brand and the second plot provides insights into the distribution of ratings across all phones in the dataset.Both plots offer valuable insights into the dataset's ratings distribution and the average ratings by brand, aiding in understanding customer satisfaction levels and potential sales performance across different brands.

```python
brand_counts = df['Brand'].value_counts()
plt.figure(figsize=(10, 6))
plt.style.use('seaborn-v0_8-white')
brand_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Distribution of Phone Brands')
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### The first plot provides insights into the distribution of phone brands in the dataset. These plots helps us to share and popularity of different brands in the dataset.

```python
dff=df.loc[df['Stars_out_of_5'] > 4.5]
px.bar(dff, x='Product Name', y='Original_Price',hover_data=['Product Name', 'Stars_out_of_5'], color='Brand',height=400,
       animation_frame='Stars_out_of_5')
```

#### It visualizes the average ROM (Read-Only Memory) for each product name, segmented by brand, with animation based on the star ratings.By animating the histogram based on star ratings, it enables the observation of how the average ROM varies across different star ratings for each product name and brand.
