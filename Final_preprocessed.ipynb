```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('flipkart_smartphones.csv')
df
```

```python
df.rename(columns = {'Title':'Product Name'}, inplace=True)
```

```python
df.dropna(subset=['Product Name'], inplace=True)
```

```python
df.reset_index(inplace = True)
```

```python
df.drop(columns='index', inplace=True)
df
```

```python
df['Color'] = df['Product Name'].str.split('(').str[1].str.split(',').str[0].str.strip()
df['Storage'] = df['Product Name'].str.split('(').str[1].str.split(',').str[1].str.split(')').str[0].str.strip()
```

```python
df
```

```python
df['Product Name'] = df['Product Name'].str.split('(').str[0].str.strip()
df
```

```python
ratings_reviews = df['Ratings'].str.split('&', expand=True)
df['Ratings'] = ratings_reviews[0].str.replace(' Ratings', '').str.replace(',', '').fillna('0').astype(int)
df['Reviews'] = ratings_reviews[1].str.replace(' Reviews', '').str.replace(',', '').str.strip().fillna('0').astype(int)
```

```python
df['Original_Price'] = df['Original_Price'].str.replace('₹', '').str.replace(',', '').fillna('0').astype(int)
df['Offer_Price'] = df['Offer_Price'].str.replace('₹', '').str.replace(',', '').fillna('0').astype(int)
df
```

```python
df = df.dropna()
df
```

```python
Discount = []

for value in df['Original_Price']:
    if value==0:
        Discount.append('No')
    else:
        Discount.append('Yes')
df['Discount'] = Discount
df
```

```python
df['Flipkart_Assured']=df['Flipkart_Assured'].replace({'Yes':1,'No':0})
```

```python
df['Color'] = df.groupby('Product Name')['Color'].transform(lambda x: ', '.join(set(x)))
df.drop_duplicates(inplace=True)
df
```

```python
df[['Brand', 'Model']] = df['Product Name'].str.split(n=1, expand=True)
df
```

```python
df['Color'] = df.groupby('Product Name')['Color'].transform(lambda x: ', '.join(set(x)))
df.drop_duplicates(inplace=True)
df
```

```python
df['5G_Support'] = df['Product Name'].apply(lambda x: 'Yes' if '5G' in x else 'No')
df
```

```python
from sklearn.preprocessing import LabelEncoder

data = np.array(df['5G_Support'])
reshaped_data = data.flatten()
label_encoder = LabelEncoder()
encoded_data = label_encoder.fit_transform(reshaped_data)
encoded_data = encoded_data.reshape(data.shape)
encoded_data
df
```

```python

label_encoder = LabelEncoder()
columns_to_encode = ['5G_Support', 'Flipkart_Assured', 'Discount']
for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])
df['Battery_mAh'] = df['Specification'].str.extract(r'(\d+)\s*mAh', expand=False)
df['Battery_mAh'] = pd.to_numeric(df['Battery_mAh'])
df['RAM'] = df['Specification'].str.extract(r'(\d+ GB RAM)')
df['ROM'] = df['Specification'].str.extract(r'(\d+ GB ROM)')
df
```

```python
df['No of Colors'] = df['Color'].apply(lambda x: len(x.split(', ')))
df
```

```python
df['RAM'] = df['RAM'].str.extract('(\d+)').fillna(0).astype(int)
df['ROM'] = df['ROM'].str.extract('(\d+)').fillna(0).astype(int)
df
```

```python
df.drop(columns=['Storage','Model'],inplace=True)
df
```

```python
df['Brand']=df['Brand'].str.lower()
battery_capacity = {
    'Apple iPhone 11': 3110,
    'Apple iPhone 12': 2815,
    'Apple iPhone 13': 3227,
    'Apple iPhone 13 Pro Max': 4352,
    'Apple iPhone 14': 3279,
    'Apple iPhone 14 Plus': 4325,
    'Apple iPhone 15': 3349,
    'Apple iPhone 15 Plus': 4383,
    'Apple iPhone 15 Pro': 3274,
    'Apple iPhone 15 Pro Max': 4422
}
for index, row in df.iterrows():
    product_name = row['Product Name']
    if pd.isna(row['Battery_mAh']) and product_name in battery_capacity:
        df.at[index, 'Battery_mAh'] = battery_capacity[product_name]
df.isna().sum()
```
