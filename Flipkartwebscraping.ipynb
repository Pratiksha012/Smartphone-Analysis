```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrape_flipkart_smartphones(num_pages):
    base_url = 'https://www.flipkart.com/search?q=smartphones'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    smartphone_data = []
    for page in range(1, num_pages + 1):
        url = f'{base_url}&page={page}'
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        products = soup.find_all('div', {'class': '_1AtVbE'})
        for product in products:
            title_elem = product.find('div', {'class': '_4rR01T'})
            title = title_elem.text.strip() if title_elem else 'N/A'

            original_price_elem = product.find('div', {'class': '_3I9_wc'})
            original_price = original_price_elem.text.strip() if original_price_elem else 'N/A'

            offer_price_elem = product.find('div', {'class': '_30jeq3'})
            offer_price = offer_price_elem.text.strip() if offer_price_elem else 'N/A'

            star_elem = product.find('div', {'class': '_3LWZlK'})
            star = star_elem.text.strip() if star_elem else 'N/A'

            rating_elem = product.find('span', {'class': '_2_R_DZ'})
            rating = rating_elem.text.strip() if rating_elem else 'N/A'
            
            flipkart_assured_elem = product.find('div', {'class': '_13J9qT'})
            flipkart_assured = 'Yes' if flipkart_assured_elem else 'No'

            description = product.find('ul', {'class':'_1xgFaf'})
            desc = description.text.strip() if description else 'N/A'
            smartphone = {
                'Title': title,
                'Original_Price': original_price,
                'Offer_Price': offer_price,
                'Stars_out_of_5': star,
                'Ratings': rating,
                'Flipkart_Assured': flipkart_assured,
                'Specification' : desc}
            smartphone_data.append(smartphone)
    return smartphone_data
```

```python
if __name__ == '__main__':
    num_pages = 1
    smartphones = scrape_flipkart_smartphones(num_pages)
    for smartphone in smartphones:
        print(smartphone)
```

```python
if __name__ == '__main__':
    num_pages = 40 
    smartphones = scrape_flipkart_smartphones(num_pages)
    df = pd.DataFrame(smartphones)
    df.to_csv('flipkart_smartphones.csv', index=False)
```
